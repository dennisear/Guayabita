<span style="color:red; font-size:30px">**D**</span>urante los años 90's era común encontrar a niños y jóvenes en las calles de nuestro país compartiendo con sus amigos y vecinos, cerraban sus calles y sacaban su balón de fútbol/baloncesto para divertirse hasta altas horas de la noche, sacaban sus juguetes o se las 'ingeniaban' creando diferentes juegos como yermis, gallinita ciega, trompo, canicas y el más popular: guayabita. Este último básicamente consiste en que cada jugador lanza un dado. Luego de obtener el resultado, apuesta a obtener un número mayor o menor al que resultó inicialmente. 

Hoy en día el casino online más importante de todo el país: el <span style="color:red">C</span>A<span style="color:red">S</span>I<span style="color:red">N</span>O <span style="color:red">L</span>U<span style="color:red">X</span>U<span style="color:red">R</span>Y <span style="color:red">P</span>L<span style="color:red">A</span>C<span style="color:red">E.</span> conocido mundialmente por sus excelentes juegos, gráficas, diversidad y métodos de pagos ha decidido celebrar su 5° aniversario brindandole un homenaje a esa niñez de los años 90's, de la mano de la empresa Globant, han lanzado oficialmente y por primera vez en el mundo digital una versión de Guayabita. En el cual se le indica al jugador los datos a ingresar y cada uno de ellos realizará un lanzamiento y se definirá el orden en el que se jugará la partida, debemos aclarar que como está en periodo de prueba empezará como máximo con 4 jugadores. A partir de este momento el juego inicia. Cada jugador tiene la posibilidad de realizar dos lanzamientos, si en el primer lanzamiento el apostador saca 1 pierde turno y también su dinero, si el apostador saca 6 gana y retira lo que apostó, en otro caso tiene la posibilidad de realizar el segundo lanzamiento y deberá escoger si se retracta de la apuesta o por el contario apuesta por un número mayor o un número menor al del primer lanzamiento y según la decisión se decide el futuro del jugador. Algo para tener en cuenta es que cada vez que un apostador retire su dinero debe recoger el 95%, pues el 5% del valor total son ganancias para el casino.

Nosotros, analistas de datos de la empresa SICEX hemos tomado la tarea de ir más a fondo e investigar la viabilidad de esta tan original idea del <span style="color:red">C</span>A<span style="color:red">S</span>I<span style="color:red">N</span>O <span style="color:red">L</span>U<span style="color:red">X</span>U<span style="color:red">R</span>Y <span style="color:red">P</span>L<span style="color:red">A</span>C<span style="color:red">E.</span> Se tiene esperado que en cada partida el casino obtenga un 5%. Hicimos un ardúo análisis y llegamos a la conclusión de que con el método **AB TESTING** podremos observar que tan viable es este juego en el casino.
